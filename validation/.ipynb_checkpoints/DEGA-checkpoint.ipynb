Import necessary packages and DEGA

````python
import pandas as pd

import DEGA
````

## Preprocessing

Import data (from the [Bottomly et al.](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0017820) dataset) about counts and phenotypes as pandas.DataFrame

````python
geneCountsFile = "bottomly_counts.csv"
phenotypeDataFile = "bottomly_phenotypes.csv"

geneCounts = pd.read_csv(geneCountsFile, index_col=0)
phenotypeData = pd.read_csv(phenotypeDataFile, index_col=0)
# filter and sort geneCounts columns on the basis of phenotypeData index
# (they have to be in the same order)
geneCounts = geneCounts[phenotypeData.index]
````

## Analysis

Load the data into a DEGA dataset

````python
dega = DEGA.dataset(geneCounts, phenotypeData, designFormula="strain")
````

Run the differential expression analysis

````python
dega.analyse()
````

## Results

A brief summary of the outcomes can be displayed with:

````python
dega.summary
````

Genes identified as up- and down-regulated can be retrieved with:

````python
dega.upregulatedGenes
````

````python
dega.downregulatedGenes
````

````python
# Save results (showAllZero is for showing in the resulting dataframe also genes with zero counts)
# dega._results(showAllZero=True).to_csv(f"DEGA_bottomly{'WithShrinkage' if dega.withShrinkage else ''}Results.csv")
````

### Plots

````python
dega.plotMA(lfcThreshold=1, highlightGenes=["ENSMUSG00000032425", "ENSMUSG00000072572"])
````

````python
dega.plotVolcano(lfcThreshold=1, labels=False, highlightGenes=["ENSMUSG00000032425", "ENSMUSG00000072572"])
````

## Further Analyses

Differential expression analysis is carried out on raw counts. However, for other downstream analyses (e.g., for visualizations or machine learning applications) it might be useful to work with transformed versions of the count data.  
DEGA provides 3 transformations: variance stabilizing transformation (default), regularized logarithm and shifted logarithm.

If counts are not transformed manually (with ```transformedCounts = dega.transformCounts()```) and provided to the following functions, DEGA will do it automatically in background.

````python
dega.plotSamplesCorrelation(lfcThreshold=1)
````

````python
dega.plotSamplesGenesCorrelation(lfcThreshold=1)
````

````python
dega.plotPCAExplainedVariance(lfcThreshold=1)
````

````python
dega.plotPCAComponentsDensity(lfcThreshold=1, n_components=3)
````

````python
dega.plotPCA(lfcThreshold=1, components=["PC1", "PC2"])
````

````python
dega.plotPCA(lfcThreshold=1, components=["PC1", "PC3"])
````

````python
dega.plotCoexpressionNetwork(lfcThreshold=1, pValueThreshold=0.01, seed=42)
````

If only the adjacency matrix is needed it can be retrieved with:

````python
dega.coexpressionNetwork(lfcThreshold=1, pValueThreshold=0.01)
````
